# <font size=5>Read results from GraphCS<font size>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
from matplotlib import cm
from matplotlib.pyplot import MultipleLocator
base='./'
savepath='simulate'
```

```python
with open('../'+savepath+'.txt','r+') as f:
    acc=[]
    for j in f:
        acc.append(float(j[:-1]))
    acc=np.array(acc).reshape((5,8))
    std=pd.read_csv(savepath+'-std.csv',header=0,index_col=0)
    std.loc['GraphCS']=np.std(acc,0)
    std.to_csv(savepath+'-std.csv')
    acc=acc.mean(0).reshape(-1)
    csv=pd.read_csv(savepath+'.csv',header=0,index_col=0)
    csv.loc['GraphCS']=acc
    csv.to_csv(savepath+'.csv')
    f.close()
```

# <font size=5>Plot Fig2a<font size>

```python
def plot_line_chart(x_names, data, x_label, y_label, line_name="", yerr=None, save_name="simulated_errorbar"):
    weight = "bold"
    font = {"size": 34}
    f, ax = plt.subplots(figsize=(15, 12))

    colors_dict = {"scmap": "#8C564B",
                   "SingleCellNet": "#92D050",
                   "Seurat-CCA": "#00B0F0",
                   "Seurat-PCA": "#F59D56",
                   "SingleR": "#BFBFBF",
                   "scGCN": "#8B008B",
                   "GraphCS": "black",
                   "CHETAH": "#CC00FF",
                   "scPred": "#FF6699",
                   'scNym':'green',
                   'scANVI':'red',
                   "CelliD(C)": "cyan",
                   "CelliD(G)": "blue",
                   "scClassify": "#E74C3C"}

    markersize = 15
    x = range(len(x_names))
    print(line_name)
    ax.set_ylim([0,1.05])
    for index in range(data.shape[0]):
        plt.errorbar(x,
                     data[index],
                     yerr=yerr[index],
                     markersize=markersize,
                     color=colors_dict[line_name[index]],
                     marker='8',
                     ecolor=colors_dict[line_name[index]],
                     elinewidth=3,
                     capsize=4,
                     linewidth=4,
                     label=line_name[index])

    plt.xticks(x, x_names, fontsize=20,weight=weight)
    plt.yticks(np.arange(0, 1.1, 0.1), fontsize=15,weight=weight)
    plt.xlabel(x_label, font, labelpad=15,fontweight=weight)
    plt.ylabel(y_label, font, labelpad=15,fontweight=weight)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_linewidth(2)
    ax.spines['left'].set_linewidth(2)
    x_major_locator = MultipleLocator(0.2)
    ax.yaxis.set_major_locator(x_major_locator)
    if weight:
        ax.spines['bottom'].set_linewidth(2)
        ax.spines['left'].set_linewidth(2)
    legend_prop = {'weight': weight, 'size': 23}
    box = ax.get_position()
    ax.set_position([box.x0, box.y0, box.width, box.height])
    if weight:
        ax.legend(loc='lower center',
                  bbox_to_anchor=(0.5,-0.4),
                  ncol=4,
                  fontsize=18,
                  prop=legend_prop
                 )
    ax.set_ylim(bottom=0.1)

    plt.tight_layout()
    plt.savefig('./figures/{}.png'.format(save_name), dpi=400, format='png', bbox_inches='tight')
    plt.show()
```

```python
df=pd.read_csv(base+'simulate-std.csv',header=0,index_col=0)
std = np.asarray(df)
df=pd.read_csv(base+'simulate.csv',header=0,index_col=0)
df.columns=[round(0.2*(i+1),1) for i in range(8)]
mean = np.asarray(df)
print(std.shape)
x_names = list(df.columns)
print(x_names)
line_name = list(df.index)
data = np.asarray(df)
x_label = "Batch.facScale"
y_label = "Accuracy"
plot_line_chart(x_names, mean, x_label, y_label, line_name=line_name, yerr=std)
```
